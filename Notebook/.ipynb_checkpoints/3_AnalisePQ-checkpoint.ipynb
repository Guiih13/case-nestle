```python
# Tratamento e Manipulao
import pandas as pd
import numpy  as np
import datetime

# Visualizao
import matplotlib.pyplot as plt
import seaborn           as sns

# Configurao
pd.set_option('max.columns', 55)
plt.style.use('bmh')
```

> # Leitura da base
---

```python
df = pd.read_csv('../Dados/Finalizados/BasePQ.csv', sep=';', parse_dates=['Data de Nascimento', 'Data de Contratacao', 'Data de Demissao'])
df.head()
```

```python
# Tratamento de colunas
df['rea'] = df['rea'].apply(lambda x: x.replace('@@@',''))
```

```python
# Passando as colunas para float

for coluna in ['Salario Base', 'Beneficios', 'VT', 'VR']:
    df[coluna] = df[coluna].astype(str)
    df[coluna] = df[coluna].apply(lambda x: x.replace(',','.'))
    df[coluna] = df[coluna].astype(float)
```

```python
df.groupby(['Nvel','rea']).agg({'CPF':pd.Series.nunique,
                                  'Salario Base':np.mean,
                                  'VR':np.mean})
```

```python
auxDf = df.groupby('Nvel')['Salario Base'].mean().reset_index()
```

```python
for coluna in ['Salario Base', 'Beneficios', 'VT', 'VR']:

    # FigSize
    plt.figure(figsize=(15,6))

    # DataFrame Auxiliar
    auxDf = df.groupby('Nvel')[coluna].mean().reset_index()

    nivel = auxDf['Nvel'] # X
    valores = auxDf[coluna] # Y

    # Grfico
    plt.bar(nivel, valores)

    # Tratamento
    plt.title(f'{coluna} x Nvel', fontweight="bold") 
    plt.xlabel('Nvel') 
    plt.ylabel(f'{coluna}')

    plt.show()
```
